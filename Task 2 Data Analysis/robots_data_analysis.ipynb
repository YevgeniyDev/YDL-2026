### Prompt
Find 3 hardly predictable correlations within this dataset

```python
# Hunt for surprising (non-obvious) correlations.
# We report BOTH Pearson and Spearman: agreement = robust; disagreement = outlier artifact.
import itertools

# All numeric columns considered for the comparison (drop id-like / bookkeeping ones)
num = df.select_dtypes(include=['float64', 'int64']).drop(
    columns=['data_curated_year', 'row_completeness_pct'], errors='ignore')
print('Columns compared (%d):' % num.shape[1])
print(list(num.columns))

def corr_pair(a, b):
    s = df[[a, b]].dropna()
    return pd.Series({
        'pearson': round(s[a].corr(s[b]), 3),
        'spearman': round(s[a].corr(s[b], method='spearman'), 3),
        'n': len(s)
    })

# --- Full pairwise scan over every numeric column pair (n >= 15) ---
scan = []
for a, b in itertools.combinations(num.columns, 2):
    cp = corr_pair(a, b)
    if cp['n'] >= 15 and pd.notna(cp['pearson']):
        scan.append({'pair': f'{a} ↔ {b}', **cp.to_dict(), 'abs_pearson': abs(cp['pearson'])})

scan = pd.DataFrame(scan).sort_values('abs_pearson', ascending=False)

# Drop the obvious / tautological pairs so only non-obvious ones remain
trivial = [
    'robot_release_year ↔ robot_age_years', 'founded_year ↔ company_age_years',
    'dof ↔ dexterity_index', 'company_age_years ↔ time_to_market_years',
    'founded_year ↔ time_to_market_years', 'max_speed_kmh ↔ walking_speed_kmh',
    'height_cm ↔ weight_kg', 'total_funding_usd_m ↔ latest_valuation_usd_b',
    'units_deployed ↔ latest_valuation_usd_b', 'units_deployed ↔ total_funding_usd_m',
    'ai_maturity_score ↔ commercial_readiness_idx',
]
non_obvious = scan[~scan['pair'].isin(trivial)]
print('\nTop non-obvious correlations:')
display(non_obvious.head(15).reset_index(drop=True))

# --- The three non-obvious correlations chosen (robust under both metrics) ---
candidates = [
    ('max_payload_kg', 'power_efficiency'),     # heavier payload -> MORE efficient
    ('walking_speed_kmh', 'max_payload_kg'),    # faster walkers also carry more
    ('company_age_years', 'unit_cost_usd_k'),   # older companies -> pricier robots
    ('speed_ratio', 'unit_cost_usd_k'),         # looks strong but is a trap (see answer)
]
result = pd.DataFrame({f'{a} ↔ {b}': corr_pair(a, b) for a, b in candidates}).T
print('\nChosen pairs:')
result
```

### Answer

3 surprising correlations (robust under **both** Pearson r and Spearman ρ):

| Correlation | r / ρ | Why it's surprising |
|---|---|---|
| payload ↔ power efficiency | +0.84 / +0.89 | heavier-rated robots are **more** efficient, not less |
| walking speed ↔ payload | +0.63 / +0.48 | fast **and** strong together (no trade-off) |
| company age ↔ cost | +0.66 / +0.43 | **older** firms build **pricier** robots |

⚠️ **Trap:** `speed_ratio ↔ cost` shows r = −0.90 but ρ = +0.11 → fake, driven by outliers. Always check Spearman against Pearson.

```python
# Helper: percent difference "a is X% more/less than b"
def pct_more(a, b):
    return round((a / b - 1) * 100, 1)

llm   = df[df['llm_integrated'] == True]
nollm = df[df['llm_integrated'] == False]

# ---------- Discovery 1: does adding an LLM make a robot more expensive? ----------
print('D1  LLM vs non-LLM unit cost (usd_k)')
print('    mean  :', round(llm['unit_cost_usd_k'].mean(),1), 'vs', round(nollm['unit_cost_usd_k'].mean(),1),
      '->', pct_more(llm['unit_cost_usd_k'].mean(),  nollm['unit_cost_usd_k'].mean()), '%')
print('    median:', llm['unit_cost_usd_k'].median(), 'vs', nollm['unit_cost_usd_k'].median(),
      '->', pct_more(llm['unit_cost_usd_k'].median(), nollm['unit_cost_usd_k'].median()), '%')
print('    check - most expensive LLM robot:',
      llm.loc[llm['unit_cost_usd_k'].idxmax(), ['robot_name','unit_cost_usd_k']].tolist(),
      '| most expensive non-LLM:', nollm.loc[nollm['unit_cost_usd_k'].idxmax(), ['robot_name','unit_cost_usd_k']].tolist())

# ---------- Discovery 2: Europe = most ready, least innovative ----------
print('\nD2  Continent innovation vs readiness')
print(df.groupby('continent')[['innovation_score','commercial_readiness_idx']].mean().round(1))
e = df[df.continent=='Europe']; na = df[df.continent=='North America']
print('    Europe vs N.America: readiness', pct_more(e['commercial_readiness_idx'].mean(), na['commercial_readiness_idx'].mean()),
      '% | innovation', pct_more(e['innovation_score'].mean(), na['innovation_score'].mean()), '%')

# ---------- Discovery 3: wheeled robots cost more than (harder) bipeds ----------
bip = df[df.biped]; whl = df[~df.biped]
print('\nD3  Wheeled vs biped unit cost')
print('    mean  :', round(whl['unit_cost_usd_k'].mean(),1), 'vs', round(bip['unit_cost_usd_k'].mean(),1),
      '->', pct_more(whl['unit_cost_usd_k'].mean(), bip['unit_cost_usd_k'].mean()), '%')
print('    median:', whl['unit_cost_usd_k'].median(), 'vs', bip['unit_cost_usd_k'].median(),
      '->', pct_more(whl['unit_cost_usd_k'].median(), bip['unit_cost_usd_k'].median()), '%   <- gap shrinks!')
print('    check - cheapest wheeled robot:', whl.loc[whl['unit_cost_usd_k'].idxmin(), ['robot_name','unit_cost_usd_k']].tolist())
```

### Findings

**1. LLMs don't make robots expensive — the cheap ones have AI.**
LLM robots are **55% cheaper** (median $67.5k vs $150k). Priciest LLM robot = Unitree H1 Pro ($110k); priciest non-LLM = Archax ($3,000k).

**2. Europe: most market-ready, least innovative.**
Europe leads readiness but is **−27% on innovation** vs North America. NA innovates most but isn't the most ready → an innovation-vs-maturity trade-off.

**3. "Wheeled costs 150% more than legged" is an outlier mirage — it even reverses.**
True by mean (+150%), but by **median wheeled is ~19% cheaper** ($72.5k vs $90k). Cheapest wheeled robot = Pepper ($25k). A few costly service/space units skew the mean.

```python
# ---- THE TRAP: Japan's $1,388k average ----
jp = df[df.country == 'Japan']
print('Japan: 6 robots, but only', jp['unit_cost_usd_k'].notna().sum(), 'have a price.')
print(jp[['robot_name', 'unit_cost_usd_k', 'deployment_stage', 'outlier_flag']].to_string(index=False))
print('\nThe "average" =', round(jp['unit_cost_usd_k'].mean()), 'k  is built from just these 4 values:',
      sorted(jp['unit_cost_usd_k'].dropna().tolist()))

# ---- THE TRUTH, one layer at a time ----
print('\n--- peel back the trap ---')
# (a) tiny n + outliers: drop the flagged outlier (Archax) and the retired museum piece (Asimo)
real = jp[(~jp.outlier_flag) & (jp.deployment_stage != 'Retired')]['unit_cost_usd_k'].dropna()
print('(a) excl. Archax (outlier_flag) + Asimo (retired):', real.tolist(),
      '-> the only priced *current* Japanese humanoid is Pepper at $25k')

# (b) missing data: 7 countries have ZERO priced robots and silently vanish from the ranking
zero = [c for c in df.country.unique() if df[df.country == c]['unit_cost_usd_k'].notna().sum() == 0]
print('(b) countries with NO price data (dropped from the ranking entirely):', zero)

# (c) duplicates: rows that are the same robot recorded twice
dups = df[df.duplicated(['company', 'height_cm', 'weight_kg', 'dof'], keep=False)]
print('(c) duplicate rows (same company/height/weight/dof) — 2 of Japan\'s 4 prices are the SAME Pepper:')
print(dups[['robot_name', 'company', 'country', 'unit_cost_usd_k', 'units_deployed']].to_string(index=False))
```

### The trap, and the truth next to it

**Claim:** "Japan builds the most expensive robots — $1,388k on average."

**Why it's a lie:** that average rests on **4 values: [25, 25, 2500, 3000]** ($k).

| Failure mode | What's really going on |
|---|---|
| **Mean over a tiny n** | n=4 (Japan has 6 robots). Two huge numbers set the average: **Asimo $2,500k** (retired since 2022 — a museum piece) and **Archax $3,000k** (a rideable mech, and it's literally `outlier_flag = True`). |
| **Held up by missing data** | 7 countries — **Canada, Norway, Germany, UK, Israel, South Korea, Iran** — have *zero* priced robots, so they silently drop out of the ranking. "Japan is #1" only among the handful that reported. |
| **Duplicate rows as "different"** | 2 of Japan's 4 prices are the **same robot**: `Pepper` and `Pepper (2025)` are identical (121 cm, 29 kg, $25k, 2000 units). Same trap hits France: `Robot One` = `Enchanted Robot One`. So "52 robots" is really **50 unique**, and Pepper's deployment is double-counted (2000 + 2000). |

**The truth:** strip the outlier and the retired piece, and the only priced *current* Japanese humanoid is **Pepper at $25k** — among the **cheapest** in the dataset, not the most expensive. The headline was exactly backwards.

**Takeaway:** before trusting a group average, always ask three questions — *how many rows? what's missing? are any rows the same thing twice?*
