# Pokémon, by the numbers — a data story

What makes one Pokémon stronger than another? With **1,072 entries** spanning eight generations, the dataset lets us answer that empirically. The thesis this notebook builds toward:

> **A Pokémon's stats aren't random — they are shaped by its _type_, its _legendary_ status, and the _generation_ it was made in — and raw stats are only half of what "strong" means, the other half being _type effectiveness_.**

Five acts make each force visible:

1. **Type identity & balance** — every type has a stat "personality", and some types are simply stronger.
2. **What makes a legendary** — legendaries are a separate power class, not just good Pokémon.
3. **Power creep across generations** — how the typical stat line, and the ceiling, moved over time.
4. **Stat archetypes & trade-offs** — the recurring builds (glass cannon, tank, all-rounder).
5. **Weaknesses & effectiveness** — the type matchup chart, and whether it offsets or compounds the stat imbalance.

…and a closing **finale** that crowns the best type combination and the strongest Pokémon.

*Tools: pandas + matplotlib/seaborn. Every figure is static and reproduces with Run All.*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('Pokemon.csv')
```

## Setup — meeting and cleaning the data

The columns above are identity (`number`, `name`, `type1`, `type2`, `generation`, `legendary`) plus the six base stats and their `total`. A data story is only as trustworthy as its data, so before analysing we fix **four issues** found in the raw file:

| # | Problem | Where | Fix |
|---|---------|-------|-----|
| 1 | `type1` = `Graass` (typo) | Eldegoss (#830) | → `Grass` |
| 2 | `type1` = `Blastoise` (a *name*, not a type) | G-max Blastoise | → `Water` |
| 3 | `total` ≠ sum of the six stats | Farfetch'd (#83): 352 vs 377 | recompute `total` |
| 4 | `generation` = 0 | Meltan #808, Melmetal #809 | → generation 7 |

We also separate **base forms** from Mega/Gigantamax variants, which share a Pokédex `number` with their base species. Most analysis uses `base` (one row per species, **898**) so types and stats aren't double-counted; the few form-specific points say so explicitly.

```python
# === Data cleaning: the four issues noted above ===

# 1 & 2. Bad values in `type1`: 'Graass' is a typo for 'Grass' (Eldegoss, #830);
#        'Blastoise' is a Pokemon *name* that leaked into the type column
#        (Gigantamax Blastoise, a Water type).
df['type1'] = df['type1'].replace({'Graass': 'Grass', 'Blastoise': 'Water'})

# 3. `total` should equal the sum of the six base stats, but Farfetch'd (#83) is
#    off (352 vs 377). Recompute it so our power measure is always trustworthy.
stats = ['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']
df['total'] = df[stats].sum(axis=1)

# 4. Two species are labelled generation 0 (Meltan #808, Melmetal #809). They
#    debuted alongside Gen 7, so place them there to keep a clean 1-8 timeline.
df['generation'] = df['generation'].replace({0: 7})

# Mega/Gigantamax and other alternate forms share a `number` with their base
# species; keeping them would double-count types and inflate stats. For species
# level analysis we keep one row per Pokedex number (the base form).
base = df.drop_duplicates('number', keep='first').copy()

# Validate every fix landed.
VALID_TYPES = {'Normal', 'Fire', 'Water', 'Electric', 'Grass', 'Ice', 'Fighting',
               'Poison', 'Ground', 'Flying', 'Psychic', 'Bug', 'Rock', 'Ghost',
               'Dragon', 'Dark', 'Steel', 'Fairy'}
assert set(df['type1']).issubset(VALID_TYPES), "Unexpected type1 value remains"
assert (df['total'] == df[stats].sum(axis=1)).all(), "total still inconsistent"
assert set(df['generation'].unique()) == set(range(1, 9)), "generation not a clean 1-8"
print(f"Cleaned OK. {len(df)} rows -> {len(base)} unique species (alternate forms dropped).")

# Official-style colour for each type, reused across every chart below.
TYPE_COLORS = {
    'Normal': '#A8A77A', 'Fire': '#EE8130', 'Water': '#6390F0',
    'Electric': '#F7D02C', 'Grass': '#7AC74C', 'Ice': '#96D9D6',
    'Fighting': '#C22E28', 'Poison': '#A33EA1', 'Ground': '#E2BF65',
    'Flying': '#A98FF3', 'Psychic': '#F95587', 'Bug': '#A6B91A',
    'Rock': '#B6A136', 'Ghost': '#735797', 'Dragon': '#6F35FC',
    'Dark': '#705746', 'Steel': '#B7B7CE', 'Fairy': '#D685AD',
}
```

## Act 1 — Type identity & balance

Pokémon are sorted into 18 elemental types, and a core design promise is that each type *feels* different — Fighting hits hard, Steel is tough, Electric is fast. Two questions:

1. **Identity** — does each type really have its own stat "personality"?
2. **Balance** — are the 18 types roughly equal in power, or are some just stronger on paper?

### 1. The census: which types are most common?

Before comparing power, look at how the roster is split. Each Pokemon counts once for its primary type and again for its secondary type (if it has one), so the bars reflect total type representation across the dex.

```python
# A Pokemon's types = type1 plus type2 (if it has one). Count every type slot.
type_counts = pd.concat([base['type1'], base['type2']]).dropna().value_counts()

fig, ax = plt.subplots(figsize=(11, 6))
ax.bar(type_counts.index, type_counts.values,
       color=[TYPE_COLORS[t] for t in type_counts.index])

ax.set_title('How common is each type?  (type1 + type2, base forms only)',
             fontsize=13, weight='bold')
ax.set_ylabel('Number of Pokemon')
ax.tick_params(axis='x', rotation=45)
for i, v in enumerate(type_counts.values):
    ax.text(i, v + 2, str(v), ha='center', va='bottom', fontsize=8)
plt.tight_layout()
plt.show()
```

### 2. The fingerprint: each type's stat personality

Averaging the six base stats within each primary type, then z-scoring per stat, reveals each type's "build" — what it leans on and what it gives up. **Blue** = above the cross-type average for that stat, **red** = below. The printed numbers are the actual mean stat values.

```python
stats = ['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']

# Average each stat within a primary type...
by_type = base.groupby('type1')[stats].mean()
# ...then z-score down each column so colour shows *relative* strength
# (which stat a type leans on), not simply who has the bigger total.
z = (by_type - by_type.mean()) / by_type.std()
z = z.sort_values('attack', ascending=False)

fig, ax = plt.subplots(figsize=(9, 9))
sns.heatmap(z, annot=by_type.loc[z.index], fmt='.0f', cmap='RdBu_r', center=0,
            linewidths=.5, cbar_kws={'label': 'relative strength (z-score)'}, ax=ax)
ax.set_title("Every type has a stat personality\n"
             "(numbers = mean stat, colour = how it compares to other types)",
             fontsize=12, weight='bold')
ax.set_xlabel('')
ax.set_ylabel('Primary type')
plt.tight_layout()
plt.show()
```

### 3. The balance check: is any type just stronger?

Stat *personality* is about shape; this is about raw size. The average base-stat total per type shows whether types cluster around the overall mean or whether some sit well above it (a fairness question for competitive play).

```python
# Is the roster balanced, or are some types just stronger on paper?
mean_total = base.groupby('type1')['total'].mean().sort_values()

fig, ax = plt.subplots(figsize=(11, 6))
ax.barh(mean_total.index, mean_total.values,
        color=[TYPE_COLORS[t] for t in mean_total.index])

overall = base['total'].mean()
ax.axvline(overall, color='black', ls='--', lw=1)
ax.text(overall + 3, 0.2, f'overall mean = {overall:.0f}', fontsize=9)

ax.set_title('Average base-stat total by primary type', fontsize=13, weight='bold')
ax.set_xlabel('Mean total stats')
plt.tight_layout()
plt.show()
```

### 4. Dual-type combinations: which pairings exist?

About half of all species (442 of 898) carry a second type. A `type1 × type2` matrix shows which combinations designers actually use — and which corners of the grid stay empty. Blank cells are pairings that never occur.

```python
# Which type1 x type2 pairings do designers actually use?
TYPE_ORDER = list(TYPE_COLORS)  # fixed 18-type order for both axes
dual = base[base['type2'].notna()]
combo = (pd.crosstab(dual['type1'], dual['type2'])
           .reindex(index=TYPE_ORDER, columns=TYPE_ORDER, fill_value=0))

fig, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(combo, annot=True, fmt='d', cmap='viridis', linewidths=.5,
            mask=(combo == 0), cbar_kws={'label': 'number of species'}, ax=ax)
ax.set_title(f'Dual-type combinations  ({len(dual)} of {len(base)} species are dual-typed)',
             fontsize=13, weight='bold')
ax.set_xlabel('Secondary type (type2)')
ax.set_ylabel('Primary type (type1)')
plt.tight_layout()
plt.show()
```

**Act 1 takeaway.** Type is a real identity, not just a label. The fingerprint heatmap shows each type leaning on a signature stat (Fighting → attack, Steel/Rock → defense, Electric → speed), the balance chart shows the roster is *not* perfectly fair (Dragon and Steel average well above the ~420 mean; Bug and Normal below it), and the combination matrix shows designers reuse a familiar handful of dual-type pairings while leaving much of the 18×18 grid empty.

## Act 2 — What makes a legendary?

Only a small share of species are flagged `legendary`. Are they simply the best ordinary Pokémon, or a different class entirely? If type sets a stat's *shape*, legendary status looks like it scales its *size*.

```python
# Distribution of total stats: legendary vs ordinary.
fig, ax = plt.subplots(figsize=(11, 6))
for flag, color, label in [(False, '#6390F0', 'Ordinary'), (True, '#EE8130', 'Legendary')]:
    subset = base[base['legendary'] == flag]['total']
    sns.kdeplot(subset, fill=True, alpha=.35, color=color, label=label, cut=0, ax=ax)
    m = subset.mean()
    ax.axvline(m, color=color, ls='--', lw=1.5)
    ax.text(m, ax.get_ylim()[1] * 0.92, f'mean {m:.0f}', color=color, ha='center', fontsize=9)

n_leg = int(base['legendary'].sum())
ax.set_title(f'Legendaries occupy a higher power tier  ({n_leg} of {len(base)} species, ~{100*n_leg/len(base):.0f}%)',
             fontsize=13, weight='bold')
ax.set_xlabel('Total base stats')
ax.legend()
plt.tight_layout()
plt.show()
```

### Where does the gap come from?

A ~190-point total gap (595 vs 405) has to live in specific stats. Comparing the mean of each of the six stats — legendary vs ordinary — shows whether legendaries are uniformly boosted or spike in particular areas.

```python
# A 190-point total gap has to live in specific stats. Compare mean per stat.
means = base.groupby('legendary')[stats].mean()

x = np.arange(len(stats)); w = 0.38
fig, ax = plt.subplots(figsize=(11, 6))
ax.bar(x - w/2, means.loc[False].values, w, label='Ordinary', color='#6390F0')
ax.bar(x + w/2, means.loc[True].values, w, label='Legendary', color='#EE8130')

for i, s in enumerate(stats):
    gap = means.loc[True, s] - means.loc[False, s]
    ax.text(i + w/2, means.loc[True, s] + 1.5, f'+{gap:.0f}',
            ha='center', fontsize=8, color='#b3541e')

ax.set_xticks(x); ax.set_xticklabels(stats, rotation=20)
ax.set_ylabel('Mean stat')
ax.set_title('Legendaries lead on all six stats — most of all on the special stats',
             fontsize=13, weight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Act 2 takeaway.** Legendaries aren't the top slice of the ordinary curve — they sit almost entirely above it (~595 vs ~405). The boost lands on *every* stat, largest on special attack (+39) and special defense (+33), so "legendary" behaves like a global power multiplier layered on top of whatever type identity a species already has.

## Act 3 — Power creep across generations

"Power creep" is the worry that newer entries are systematically stronger. With eight generations recorded, we can check it directly — the average per generation, the spread behind that average, and what Mega/Gigantamax forms did to the ceiling.

```python
# Mean total per generation (base forms; gen 0 already folded into gen 7).
gen = base.groupby('generation')['total'].mean()

fig, ax = plt.subplots(figsize=(11, 6))
ax.plot(gen.index, gen.values, marker='o', lw=2, color='#6F35FC')
for g, v in gen.items():
    ax.annotate(f'{v:.0f}', (g, v), textcoords='offset points', xytext=(0, 9),
                ha='center', fontsize=8)
overall = base['total'].mean()
ax.axhline(overall, color='grey', ls='--', lw=1, label=f'overall mean {overall:.0f}')

ax.set_xlabel('Generation')
ax.set_ylabel('Mean total base stats')
ax.set_title('Average base-stat total by generation', fontsize=13, weight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

### Means hide the spread

A flat average can still conceal widening extremes. A box plot per generation shows the full distribution — medians, spread, and the strong outliers each generation introduced.

```python
# A flat average can still hide widening extremes, so show the full spread per gen.
fig, ax = plt.subplots(figsize=(11, 6))
sns.boxplot(data=base, x='generation', y='total', hue='generation',
            palette='viridis', legend=False, ax=ax)
ax.set_xlabel('Generation')
ax.set_ylabel('Total base stats')
ax.set_title('Distribution of total stats per generation (base forms)',
             fontsize=13, weight='bold')
plt.tight_layout()
plt.show()
```

### The real ceiling: Mega & Gigantamax forms

Base forms tell only part of the story. Bringing back the alternate forms we set aside, we can compare the strongest *base* form against the strongest form overall in each generation — the gap is where the franchise's true power ceiling lives.

```python
# Base forms tell only part of the story. Bring back the alternate forms (full df)
# and compare the strongest base form vs the strongest form overall, per generation.
gens = sorted(df['generation'].unique())
base_max = base.groupby('generation')['total'].max().reindex(gens)
all_max = df.groupby('generation')['total'].max().reindex(gens)

x = np.arange(len(gens)); w = 0.38
fig, ax = plt.subplots(figsize=(11, 6))
ax.bar(x - w/2, base_max.values, w, label='Strongest base form', color='#B7B7CE')
ax.bar(x + w/2, all_max.values, w, label='Strongest incl. Mega / G-max', color='#C22E28')

for i, (b, a) in enumerate(zip(base_max.values, all_max.values)):
    if a > b:
        ax.text(i + w/2, a + 6, f'+{a - b:.0f}', ha='center', fontsize=8, color='#8c1f1b')

ax.set_xticks(x); ax.set_xticklabels(gens)
ax.set_xlabel('Generation')
ax.set_ylabel('Maximum total base stats')
ax.set_title('Alternate forms lifted the power ceiling (Gen 8 reaches 1125)',
             fontsize=13, weight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Act 3 takeaway.** Power creep in the *typical* Pokémon is mild: generation means drift from ~405 in the early gens to ~430–450 later, with no runaway climb. The real escalation is at the top — Mega and especially Gigantamax forms push the maximum far beyond any base form, peaking at **1125** in Gen 8. That gap between a stable middle and an exploding ceiling is what "power creep" usually refers to.

## Act 4 — Stat archetypes & trade-offs

A Pokémon has a finite "stat budget" — its `total` — so in principle raising one stat means lowering another. This act looks at how the six stats actually relate, and at the recurring builds those relationships produce.

```python
# How do the six stats (and the total) move together?
corr = base[stats + ['total']].corr()

fig, ax = plt.subplots(figsize=(8, 7))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu_r', center=0, vmin=-1, vmax=1,
            linewidths=.5, square=True, cbar_kws={'label': 'correlation'}, ax=ax)
ax.set_title('How the six stats relate (base forms)', fontsize=13, weight='bold')
plt.tight_layout()
plt.show()
```

### Glass cannons vs tanks

Plotting attack against defense splits the roster along its most intuitive trade-off: high-attack / low-defense **glass cannons** in one corner, low-attack / high-defense **tanks** in the other. Point size tracks total power and colour flags legendaries.

```python
# Attack vs defense splits the roster along its most intuitive trade-off.
ordinary = base[~base['legendary']]
legendary = base[base['legendary']]

fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(ordinary['attack'], ordinary['defense'], s=ordinary['total'] / 8,
           alpha=.30, color='#6390F0', edgecolor='none', label='Ordinary')
ax.scatter(legendary['attack'], legendary['defense'], s=legendary['total'] / 8,
           alpha=.65, color='#EE8130', edgecolor='none', label='Legendary')

# Median guides split the plane into archetype quadrants.
ax.axvline(base['attack'].median(), color='grey', ls=':', lw=1)
ax.axhline(base['defense'].median(), color='grey', ls=':', lw=1)
ax.text(0.98, 0.02, 'glass cannon →\nhigh attack, low defense', transform=ax.transAxes,
        ha='right', va='bottom', fontsize=9, color='#444')
ax.text(0.02, 0.98, '← tank\nlow attack, high defense', transform=ax.transAxes,
        ha='left', va='top', fontsize=9, color='#444')

ax.set_xlabel('Attack')
ax.set_ylabel('Defense')
ax.set_title('Attack vs defense: the archetype plane\n(point size = total stats, colour = legendary)',
             fontsize=12, weight='bold')
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

**Act 4 takeaway.** At the population level no stat trades off *against* another — attack and defense are even mildly *positive* (r ≈ 0.46), because high-`total` Pokémon have more of everything. The trade-offs live *within* the budget: the attack–defense plane still spreads species into recognisable glass-cannon and tank corners. And "strong" here mostly means "hits hard" — the offensive stats (attack, sp_attack, r ≈ 0.72) track `total` more closely than the defensive ones.

## Act 5 — Weaknesses & effectiveness: the other half of balance

Act 1 ranked types by raw stats and called the roster *unfair* — Dragon and Steel on top, Bug and Normal at the bottom. But stats are only half of balance: in battle every move is scaled by the **type effectiveness chart**, which rewards or punishes a Pokémon regardless of its stat line. That matchup data isn't in the CSV, so we bring in the canonical 18-type chart and ask the obvious follow-up — does typing *offset* the stat imbalance, or *compound* it?

> Multipliers: **×2** super-effective, **×½** resisted, **×0** no effect (×1 neutral). A dual-type Pokémon multiplies its two types, so real-world weaknesses span ×0 to ×4.

```python
# === The canonical Gen 6+ type chart (NOT in the CSV — external game data) ===
# chart.loc[attacker, defender] = damage multiplier (0, 0.5, 1 or 2).
# Only the non-neutral matchups are listed; everything else defaults to 1.
TYPE_ORDER = list(TYPE_COLORS)  # the 18 types, fixed order

_SUPER = {  # x2  (super-effective)
    'Normal': [], 'Fire': ['Grass', 'Ice', 'Bug', 'Steel'],
    'Water': ['Fire', 'Ground', 'Rock'], 'Electric': ['Water', 'Flying'],
    'Grass': ['Water', 'Ground', 'Rock'], 'Ice': ['Grass', 'Ground', 'Flying', 'Dragon'],
    'Fighting': ['Normal', 'Ice', 'Rock', 'Dark', 'Steel'], 'Poison': ['Grass', 'Fairy'],
    'Ground': ['Fire', 'Electric', 'Poison', 'Rock', 'Steel'], 'Flying': ['Grass', 'Fighting', 'Bug'],
    'Psychic': ['Fighting', 'Poison'], 'Bug': ['Grass', 'Psychic', 'Dark'],
    'Rock': ['Fire', 'Ice', 'Flying', 'Bug'], 'Ghost': ['Psychic', 'Ghost'], 'Dragon': ['Dragon'],
    'Dark': ['Psychic', 'Ghost'], 'Steel': ['Ice', 'Rock', 'Fairy'], 'Fairy': ['Fighting', 'Dragon', 'Dark'],
}
_NOTVERY = {  # x1/2 (resisted)
    'Normal': ['Rock', 'Steel'], 'Fire': ['Fire', 'Water', 'Rock', 'Dragon'],
    'Water': ['Water', 'Grass', 'Dragon'], 'Electric': ['Electric', 'Grass', 'Dragon'],
    'Grass': ['Fire', 'Grass', 'Poison', 'Flying', 'Bug', 'Dragon', 'Steel'],
    'Ice': ['Fire', 'Water', 'Ice', 'Steel'], 'Fighting': ['Poison', 'Flying', 'Psychic', 'Bug', 'Fairy'],
    'Poison': ['Poison', 'Ground', 'Rock', 'Ghost'], 'Ground': ['Grass', 'Bug'],
    'Flying': ['Electric', 'Rock', 'Steel'], 'Psychic': ['Psychic', 'Steel'],
    'Bug': ['Fire', 'Fighting', 'Poison', 'Flying', 'Ghost', 'Steel', 'Fairy'],
    'Rock': ['Fighting', 'Ground', 'Steel'], 'Ghost': ['Dark'], 'Dragon': ['Steel'],
    'Dark': ['Fighting', 'Dark', 'Fairy'], 'Steel': ['Fire', 'Water', 'Electric', 'Steel'],
    'Fairy': ['Fire', 'Poison', 'Steel'],
}
_NOEFFECT = {  # x0  (immune)
    'Normal': ['Ghost'], 'Electric': ['Ground'], 'Fighting': ['Ghost'], 'Poison': ['Steel'],
    'Ground': ['Flying'], 'Psychic': ['Dark'], 'Ghost': ['Normal'], 'Dragon': ['Fairy'],
}

chart = pd.DataFrame(1.0, index=TYPE_ORDER, columns=TYPE_ORDER)
for atk in TYPE_ORDER:
    for d in _SUPER.get(atk, []):    chart.loc[atk, d] = 2.0
    for d in _NOTVERY.get(atk, []):  chart.loc[atk, d] = 0.5
    for d in _NOEFFECT.get(atk, []): chart.loc[atk, d] = 0.0

assert chart.shape == (18, 18)
assert set(pd.unique(chart.values.ravel())) <= {0.0, 0.5, 1.0, 2.0}
print('Type chart built:', chart.shape,
      '| multipliers:', sorted(set(pd.unique(chart.values.ravel()))))
```

### The matchup chart

Reading a *row* gives a type's offense (what its moves hit hard); reading a *column* gives its defense (what hits it hard). The grid isn't symmetric — some types are built to attack, others to endure.

```python
import matplotlib.colors as mcolors
from matplotlib.patches import Patch

# Each move is scaled by the type chart, so colour the four multiplier levels.
cmap = mcolors.ListedColormap(['#3a3a3a', '#e8896b', '#f2f2f2', '#79c879'])  # 0, 1/2, 1, 2
norm = mcolors.BoundaryNorm([-0.25, 0.25, 0.75, 1.5, 2.5], cmap.N)
annot = chart.replace({0.0: '0', 0.5: '½', 1.0: '', 2.0: '×2'})

fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(chart, cmap=cmap, norm=norm, annot=annot, fmt='', cbar=False,
            linewidths=.5, linecolor='white', annot_kws={'fontsize': 8}, ax=ax)
ax.set_title('Type effectiveness chart — attacking (rows) vs defending (columns)',
             fontsize=13, weight='bold')
ax.set_xlabel('Defending type')
ax.set_ylabel('Attacking type')
ax.tick_params(axis='x', rotation=90)
ax.tick_params(axis='y', rotation=0)
ax.legend(handles=[Patch(color='#79c879', label='×2 super-effective'),
                   Patch(color='#f2f2f2', ec='grey', label='×1 neutral'),
                   Patch(color='#e8896b', label='×½ resisted'),
                   Patch(color='#3a3a3a', label='×0 no effect')],
          bbox_to_anchor=(1.01, 1), loc='upper left', frameon=False)
plt.tight_layout()
plt.show()
```

### Offensive reach

Counting the ×2 cells along each attacker's row measures coverage: how many of the 18 types it threatens. This is pure typing — completely independent of a Pokémon's stats.

```python
# Read ACROSS each attacker's row: how many types it hits super-effectively (x2).
offense = (chart == 2).sum(axis=1).sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(11, 6))
ax.bar(offense.index, offense.values,
       color=[TYPE_COLORS[t] for t in offense.index])
for i, v in enumerate(offense.values):
    ax.text(i, v + 0.06, str(int(v)), ha='center', fontsize=8)
ax.set_title('Offensive reach: how many of the 18 types each attack hits super-effectively',
             fontsize=12, weight='bold')
ax.set_ylabel('# types hit for ×2')
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

### Defensive resilience

Reading *down* each defender's column gives its weaknesses (×2 taken), resistances (×½) and immunities (×0). The net — resistances plus immunities minus weaknesses — is a single type's defensive score. (Real dual-type Pokémon stack two columns, so their multipliers run ×0 to ×4.)

```python
# Read DOWN each defender's column: weaknesses (x2 taken), resistances (x1/2), immunities (x0).
weak = (chart == 2).sum(axis=0)
resist = (chart == 0.5).sum(axis=0)
immune = (chart == 0).sum(axis=0)
prof = pd.DataFrame({'weak': weak, 'resist': resist, 'immune': immune})
prof['net'] = prof['resist'] + prof['immune'] - prof['weak']
prof = prof.sort_values('net')

fig, ax = plt.subplots(figsize=(11, 7.5))
ax.barh(prof.index, -prof['weak'], color='#d24c46', label='weaknesses (×2 taken)')
ax.barh(prof.index, prof['resist'], color='#5aa469', label='resistances (×½)')
ax.barh(prof.index, prof['immune'], left=prof['resist'], color='#3a6ea5', label='immunities (×0)')
ax.axvline(0, color='black', lw=.8)
for t, row in prof.iterrows():
    ax.text(row['resist'] + row['immune'] + 0.15, t, f"net {row['net']:+.0f}",
            va='center', fontsize=8)
ax.set_title('Defensive typing: resistances & immunities vs weaknesses (single-type)',
             fontsize=12, weight='bold')
ax.set_xlabel('count   (← weaknesses        resistances / immunities →)')
ax.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

### Back to the controversy: does typing rebalance the roster?

Act 1's imbalance was about *stats*. If type effectiveness were a counter-balancing lever, the weaker-stat types would receive *better* typing to compensate — a downward trend. Plotting each type's mean stat total against its offensive reach and its defensive score tests that directly.

```python
# Does typing offset the stat imbalance from Act 1? Compare each type's mean stat
# total against its OFFENSIVE reach and its DEFENSIVE score.
mean_total = base.groupby('type1')['total'].mean()
off_se = (chart == 2).sum(axis=1)                                   # types hit x2
def_net = (chart == 0.5).sum(axis=0) + (chart == 0).sum(axis=0) - (chart == 2).sum(axis=0)
comp = pd.DataFrame({'mean_total': mean_total, 'def_net': def_net, 'off_se': off_se}).dropna()

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
panels = [('def_net', 'Defensive net  (resist + immune − weak)'),
          ('off_se', 'Offensive reach  (types hit ×2)')]
for ax, (ycol, ylab) in zip(axes, panels):
    ax.scatter(comp['mean_total'], comp[ycol], s=90,
               c=[TYPE_COLORS[t] for t in comp.index], edgecolor='grey', zorder=3)
    for t, r in comp.iterrows():
        ax.annotate(t, (r['mean_total'], r[ycol]), fontsize=7,
                    xytext=(4, 3), textcoords='offset points')
    m, b = np.polyfit(comp['mean_total'], comp[ycol], 1)
    xs = np.array([comp['mean_total'].min(), comp['mean_total'].max()])
    ax.plot(xs, m * xs + b, ls='--', color='grey', lw=1.2)
    rho = comp['mean_total'].corr(comp[ycol])
    ax.set_xlabel('Mean total base stats (by primary type)')
    ax.set_ylabel(ylab)
    ax.set_title(f'r = {rho:.2f}', fontsize=11)
fig.suptitle('Does type effectiveness offset the stat imbalance from Act 1?',
             fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

**Act 5 takeaway.** Typing does **not** rescue the stat underdogs — defensively it *compounds* the imbalance: mean stats and defensive score are positively correlated (r ≈ 0.33), with **Steel** the extreme — top-tier stats *and* the best defensive typing (8 resistances/immunities, zero net weaknesses). But the Act 1 controversy is partly resolved on **offense**: stat strength and offensive reach are *uncorrelated* (r ≈ 0), and **Dragon** is the proof — second-highest stats yet the worst coverage (super-effective on just one type) and, since Gen 6, a hard Fairy counter. So balance is real but unevenly applied: Steel is strong everywhere, Dragon is a paper giant, and frail types like Bug/Ice are punished on both fronts — 172 of 898 species even carry a crippling ×4 weakness.

## Finale — the optimum: best typing & strongest Pokémon

The acts mapped the forces; this finale names winners. Two questions:

1. **What is the best *type combination*?** Using the matchup chart from Act 5, we score all **171** possible typings (18 single + 153 dual) on **defence** (resistances + immunities − weaknesses, with dual-types stacking ×0–×4) and **offence** (how many of the 18 types its STAB hits super-effectively).
2. **Which is the strongest *Pokémon* right now?** Ranked by base-stat `total` — both overall (alternate forms included) and among base species.

```python
import itertools

# Score every possible typing on the matchup chart from Act 5.
def summarise(types):
    """Defensive multipliers this 1- or 2-type combo takes, plus offensive reach."""
    mult = pd.Series(1.0, index=TYPE_ORDER)
    for atk in TYPE_ORDER:
        m = 1.0
        for t in types:
            m *= chart.loc[atk, t]
        mult[atk] = m
    weak = int((mult > 1).sum())
    resist = int(((mult < 1) & (mult > 0)).sum())
    immune = int((mult == 0).sum())
    quad = int((mult >= 4).sum())                                  # crippling x4 weaknesses
    off = sum(any(chart.loc[t, d] == 2 for t in types) for d in TYPE_ORDER)
    return dict(weak=weak, resist=resist, immune=immune, quad=quad,
                def_net=resist + immune - weak, off_cov=off)

combos = [(t,) for t in TYPE_ORDER] + list(itertools.combinations(TYPE_ORDER, 2))
combo_df = pd.DataFrame([{'combo': '/'.join(c), 'types': c, **summarise(c)} for c in combos])
print(len(combo_df), 'possible typings  (18 single + 153 dual)')

best_def = combo_df.sort_values(['def_net', 'weak', 'quad'], ascending=[False, True, True]).iloc[0]
best_off = combo_df.sort_values('off_cov', ascending=False).iloc[0]
print(f"Best defensive typing : {best_def['combo']:14s}  net {best_def['def_net']:+d}  "
      f"({best_def['resist']} resist, {best_def['immune']} immune, {best_def['weak']} weak)")
print(f"Widest offensive reach: {best_off['combo']:14s}  hits {best_off['off_cov']} of 18 types ×2")
```

### Best defensive typing

A dual type multiplies its two columns, so weaknesses and resistances *stack* (×0 to ×4). Scoring every combination by net resistances surfaces one clear winner — and a striking pattern in the runners-up.

```python
# Rank the most defensively resilient combos (stacked weaknesses vs resist/immune).
top = combo_df.sort_values(['def_net', 'weak', 'quad'],
                           ascending=[False, True, True]).head(12).iloc[::-1]

fig, ax = plt.subplots(figsize=(11, 7))
ax.barh(top['combo'], -top['weak'], color='#d24c46', label='weaknesses (×2/×4 taken)')
ax.barh(top['combo'], top['resist'], color='#5aa469', label='resistances')
ax.barh(top['combo'], top['immune'], left=top['resist'], color='#3a6ea5', label='immunities')
ax.axvline(0, color='black', lw=.8)
for combo, row in top.set_index('combo').iterrows():
    tag = f"net {int(row['def_net']):+d}" + (f"   ⚠ {int(row['quad'])}× ×4" if row['quad'] else '')
    ax.text(row['resist'] + row['immune'] + 0.2, combo, tag, va='center', fontsize=8)
ax.set_title('The most defensively resilient type combinations', fontsize=13, weight='bold')
ax.set_xlabel('count   (← weaknesses        resistances / immunities →)')
ax.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

### The whole map: offence vs defence

A great defensive wall that can't threaten anything is only half a Pokémon. Plotting all 171 typings by offensive coverage *and* defensive net shows the trade-off space — and how few combinations are strong in both directions at once.

```python
# Map every typing by offence vs defence; the corners are the interesting ones.
fig, ax = plt.subplots(figsize=(11, 8))
ax.scatter(combo_df['off_cov'], combo_df['def_net'], s=28, color='#bbbbbb',
           alpha=.6, zorder=2, label='all 171 typings')

flagged = {'Steel/Fairy': 'best defence', 'Fire/Ground': 'best all-round',
           'Ice/Ground': 'best offence', 'Steel': 'Steel alone'}
for name, note in flagged.items():
    r = combo_df[combo_df['combo'] == name].iloc[0]
    ax.scatter(r['off_cov'], r['def_net'], s=140, color='#cc3333',
               edgecolor='black', zorder=3)
    ax.annotate(f"{name}\n({note})", (r['off_cov'], r['def_net']),
                fontsize=8, weight='bold', xytext=(7, 4), textcoords='offset points')

ax.axhline(0, color='grey', ls=':', lw=1)
ax.set_xlabel('Offensive coverage  (types hit ×2 by either STAB)')
ax.set_ylabel('Defensive net  (resistances + immunities − weaknesses)')
ax.set_title('The typing map: all 171 combinations\n(up = tougher to hit, right = wider coverage)',
             fontsize=12, weight='bold')
ax.legend(loc='lower left')
plt.tight_layout()
plt.show()
```

### The strongest Pokémon

"Strongest" means the highest base-stat `total` — but a Mega/Gigantamax form isn't comparable to a wild encounter, and legendaries are their own class. So we crown **four** separate champions, splitting the roster by *base vs alternate form* and *regular vs legendary*: the best regular Pokémon, the best legendary, the best alternate form, and the best alternate legendary.

```python
# Four kinds of "strongest" on a 2x2 grid. An alternate form = any non-first row
# sharing a Pokedex number (Mega / Gigantamax / Primal / special forme).
df['is_alt'] = df.duplicated('number', keep='first')

categories = [
    ('Best regular Pokémon  (base form, non-legendary)', (~df['is_alt']) & (~df['legendary'])),
    ('Best legendary  (base form)',                       (~df['is_alt']) & (df['legendary'])),
    ('Best alternate form  (non-legendary)',              (df['is_alt']) & (~df['legendary'])),
    ('Best alternate legendary',                          (df['is_alt']) & (df['legendary'])),
]

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
for ax, (title, mask) in zip(axes.ravel(), categories):
    d = df[mask].sort_values('total', ascending=False).head(6).iloc[::-1]
    bars = ax.barh(d['name'], d['total'], color=[TYPE_COLORS[t] for t in d['type1']])
    bars[-1].set_edgecolor('black'); bars[-1].set_linewidth(2)   # emphasise the champion (top bar)
    for nm, tot in zip(d['name'], d['total']):
        ax.text(tot - 8, nm, str(int(tot)), va='center', ha='right',
                fontsize=8, color='white', weight='bold')
    champ = d.iloc[-1]
    ax.set_title(f"{title}\n→ {champ['name']} ({int(champ['total'])})", fontsize=11, weight='bold')
    ax.set_xlabel('Total base stats')
fig.suptitle('The strongest Pokémon, by category  (bar colour = primary type)',
             fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

**Finale answers.**

- **Best type combination — defensively: `Steel / Fairy`.** Nine resistances, two immunities, only two weaknesses and *no* ×4 — the most resilient typing in the game. Steel is the defensive backbone of *every* top combo. Real examples: **Mawile, Klefki, Magearna**.
- **Best offensively:** Ground-based combos hit widest — `Ice / Ground` and `Ice / Fighting` are super-effective against **9 of 18** types, though they pay for it with fragile defence.
- **Best all-rounder: `Fire / Ground`** (Primal Groudon's typing) — wide coverage *and* a positive defensive score.

- **Strongest Pokémon, by category:**
  - **Regular (base, non-legendary): Slaking — 670** (next best are a cluster of 600s: Dragapult, Garchomp, Metagross…).
  - **Legendary (base): Arceus — 720** (then Eternatus 690, and a wall of 680s).
  - **Alternate form (non-legendary): a four-way tie at 700** — Mega Garchomp, Mega Metagross, Mega Salamence, Mega Tyranitar.
  - **Alternate legendary: Eternamax Eternatus — 1125**, the single highest total in the entire dataset.

## Conclusion

The design forces, one coherent picture:

- **Type** sets the *shape* of a Pokémon's stats — each type has a measurable personality, and the 18 types are deliberately *un*equal in raw power (Dragon and Steel sit well above the ~420 average).
- **Legendary status** scales the *size* — a near-uniform boost across all six stats that lifts a species into its own tier (~595 vs ~405), led by special attack.
- **Generation** kept the *typical* Pokémon fairly stable (~405 → ~450 across eight gens — drift, not runaway creep), while Mega/Gigantamax forms quietly raised the ceiling to 1125.
- **Type effectiveness** is the *other half* of balance, and it doesn't level the field: defensively it compounds the stat gap (Steel is strong **and** near-unbreakable), while offensively it humbles a few stat giants — Dragon's raw power comes with the worst type coverage and a Fairy hard-counter.

So a stat line is far from arbitrary: knowing a Pokémon's type, whether it's legendary, when it was made, and how its typing plays on the matchup chart already explains most of what makes it strong — and how strong it *actually* is in battle. The data tells a *designed* story.
