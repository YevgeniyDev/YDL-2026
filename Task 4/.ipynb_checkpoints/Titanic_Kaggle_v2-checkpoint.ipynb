# Titanic — v2: pushing a Linear Regression *honestly* past 0.80

v1 reached **0.80143** on Kaggle with `Family_Survival` + a lean, binned feature set.
This notebook (separate from v1 — `Titanic_Kaggle.ipynb` is untouched) tries to squeeze a
bit more, **still using only plain `LinearRegression`**. Two genuinely new ideas:

1. **Interaction features.** A linear model fits a flat plane; it *cannot* represent that
   class matters far more for women than men. The data is stark — 1st-class women survived
   97%, 3rd-class women only 50%, while across men the class effect is much smaller. We give
   the line the curvature it needs by adding explicit **`Sex×Pclass`** and **`Sex×AgeBin`**
   terms. This is the single biggest lever available *specifically to a linear model*.
2. **New honest features.** `HasCabin` (a recorded cabin ⇒ 0.67 survival vs 0.30 without),
   and `TicketGroupSize` (people sharing a ticket — small/mid groups lived, big groups died).

Plus a methodology fix:

3. **Leak-free cross-validation.** v1's CV (~0.85) was *optimistic* because `Family_Survival`
   peeked at all training labels. Here we recompute it **inside each fold from that fold's
   training labels only**, so the CV number is finally trustworthy and we can compare feature
   sets on solid ground.

Output is written to **`submission1.csv`**.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score

train = pd.read_csv('train.csv')
test  = pd.read_csv('test.csv')
n_train = len(train)
y = train['Survived'].astype(float)
print('train:', train.shape, ' test:', test.shape)
```

## 1. Base feature engineering (no labels used)

Everything here is a deterministic transform of the raw columns — no `Survived` is touched,
so it is safe to compute once on the combined train+test frame. `Family_Survival` is the
only label-derived feature and is handled separately (and leak-safely) below.

```python
combined = pd.concat([train, test], sort=False, ignore_index=True)

# Title from name
combined['RawTitle'] = combined['Name'].str.extract(r',\s*([^.]+)\.')[0].str.strip()
title_map = {
    'Mr':'Mr','Mrs':'Mrs','Miss':'Miss','Master':'Master','Ms':'Miss','Mlle':'Miss',
    'Mme':'Mrs','Lady':'Rare','Countess':'Rare','Dona':'Rare','Don':'Rare','Sir':'Rare',
    'Jonkheer':'Rare','Dr':'Rare','Rev':'Rare','Col':'Rare','Major':'Rare','Capt':'Rare',
}
combined['Title'] = combined['RawTitle'].map(title_map).fillna('Rare')

combined['Sex_female'] = (combined['Sex'] == 'female').astype(int)
combined['FamilySize'] = combined['SibSp'] + combined['Parch'] + 1
combined['IsAlone']    = (combined['FamilySize'] == 1).astype(int)

# Imputation
combined['Age']  = combined['Age'].fillna(combined.groupby('Title')['Age'].transform('median'))
combined['Fare'] = combined['Fare'].fillna(combined.groupby('Pclass')['Fare'].transform('median'))

# Bins
combined['AgeBin']  = pd.cut(combined['Age'], bins=[0,14,25,35,45,60,200], labels=False)
combined['FareBin'] = pd.qcut(combined['Fare'], 5, labels=False)

# NEW honest features
combined['HasCabin']        = combined['Cabin'].notna().astype(int)
combined['TicketGroupSize'] = combined.groupby('Ticket')['Ticket'].transform('count')

# NEW interaction terms (give the linear model curvature it otherwise cannot represent)
combined['Sex_x_Pclass'] = combined['Sex_female'] * combined['Pclass']
combined['Sex_x_AgeBin'] = combined['Sex_female'] * combined['AgeBin']

# One-hot Title (keep Master = young boys)
combined = pd.get_dummies(combined, columns=['Title'], dtype=int)

# Needed for the group feature
combined['LastName'] = combined['Name'].str.split(',').str[0]

print('Base features ready. combined:', combined.shape)
```

## 2. Leak-safe `Family_Survival`

The function reads `Survived` **only for rows we explicitly allow** (`known_mask`). For
cross-validation we pass the fold's *training* rows; for the final model we pass *all*
training rows. A passenger's own label is always excluded (`drop(idx)`). Test labels never
exist, so they are never used.

```python
def family_survival(df, known_mask):
    surv = df['Survived'].where(pd.Series(known_mask, index=df.index))
    fs = pd.Series(0.5, index=df.index)
    # (a) same surname + fare
    for _, grp in df.groupby(['LastName', 'Fare']):
        if len(grp) > 1:
            idxs = grp.index
            for idx in idxs:
                others = surv.loc[idxs.drop(idx)]
                if others.max() == 1.0:   fs.loc[idx] = 1.0
                elif others.min() == 0.0: fs.loc[idx] = 0.0
    # (b) fall back to same ticket
    for _, grp in df.groupby('Ticket'):
        if len(grp) > 1:
            idxs = grp.index
            for idx in idxs:
                if fs.loc[idx] in (0.0, 0.5):
                    others = surv.loc[idxs.drop(idx)]
                    if others.max() == 1.0:   fs.loc[idx] = 1.0
                    elif others.min() == 0.0: fs.loc[idx] = 0.0
    return fs
```

## 3. Honest cross-validation harness (multi-seed)

For each fold we precompute a **fold-safe** `Family_Survival` from that fold's training
labels only — no cross-fold leakage. We repeat over **5 random seeds** and average, because a
single split makes the best threshold jump around (one split here happened to pick 0.60,
which is just noise). Averaging gives a trustworthy accuracy *and* a stable threshold.

```python
SEEDS = range(5)

# Precompute fold-safe Family_Survival for every (seed, fold) once.
seed_folds = {}
for seed in SEEDS:
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
    per_fold = []
    for tr_idx, va_idx in skf.split(np.arange(n_train), y):
        known = np.zeros(len(combined), dtype=bool)
        known[tr_idx] = True              # only this fold's TRAIN rows are 'known'
        per_fold.append((tr_idx, va_idx, family_survival(combined, known)))
    seed_folds[seed] = per_fold

def honest_oof(feature_cols, seed):
    oof = np.zeros(n_train)
    for tr_idx, va_idx, fs in seed_folds[seed]:
        combined['Family_Survival'] = fs.values
        X = combined[feature_cols].astype(float)
        model = LinearRegression().fit(X.iloc[tr_idx], y.iloc[tr_idx])
        oof[va_idx] = model.predict(X.iloc[va_idx])
    return oof

THRESHOLDS = np.arange(0.30, 0.71, 0.01)

def eval_set(feature_cols):
    # Average the accuracy-vs-threshold curve across seeds, then pick the peak.
    acc = np.zeros(len(THRESHOLDS))
    for seed in SEEDS:
        oof = honest_oof(feature_cols, seed)
        acc += [accuracy_score(y, (oof >= t).astype(int)) for t in THRESHOLDS]
    acc /= len(SEEDS)
    i = int(acc.argmax())
    return THRESHOLDS[i], acc[i]
```

## 4. Compare feature sets on the honest CV

We test the v1 set against v1 + interactions, and v1 + interactions + the new cabin/ticket
features. The leak-free CV decides which actually generalises — no guessing.

```python
sets = {
    'v1 (lean, 7)':           ['Family_Survival','Sex_female','Pclass','AgeBin','FareBin','FamilySize','Title_Master'],
    'v1 + interactions':      ['Family_Survival','Sex_female','Pclass','AgeBin','FareBin','FamilySize','Title_Master',
                               'Sex_x_Pclass','Sex_x_AgeBin'],
    'v1 + inter + cabin/grp': ['Family_Survival','Sex_female','Pclass','AgeBin','FareBin','FamilySize','Title_Master',
                               'Sex_x_Pclass','Sex_x_AgeBin','HasCabin','TicketGroupSize'],
}

results = {}
for name, cols in sets.items():
    t, acc = eval_set(cols)
    results[name] = (cols, t, acc)
    print(f'{name:26s}  honest CV acc = {acc:.4f}  (threshold {t:.2f})')

# Prefer the simplest set within 0.001 of the best accuracy (fewer features generalise better)
top = max(r[2] for r in results.values())
best_name = next(n for n, c in sets.items() if results[n][2] >= top - 0.001)
best_cols, best_t, best_acc = results[best_name]
print(f'\nChosen (simplest near-best): {best_name}  ->  {best_acc:.4f} at threshold {best_t:.2f}')
```

## 5. Inspect the chosen model's coefficients

```python
known_full = np.zeros(len(combined), dtype=bool)
known_full[:n_train] = True
combined['Family_Survival'] = family_survival(combined, known_full).values   # all train labels

X      = combined.iloc[:n_train][best_cols].astype(float)
X_test = combined.iloc[n_train:][best_cols].astype(float).reset_index(drop=True)

insp = LinearRegression().fit(X, y)
print(pd.Series(insp.coef_, index=best_cols).sort_values())
```

## 6. Train on all data, write `submission1.csv`

```python
final = LinearRegression().fit(X, y)
scores = final.predict(X_test)
pred = (scores >= best_t).astype(int)

submission = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': pred})
submission.to_csv('submission1.csv', index=False)
print('Wrote submission1.csv', submission.shape)
print('Predicted survivors:', int(pred.sum()), 'of', len(pred), f'({pred.mean():.1%})')
submission.head()
```

## Summary — what we learned (the honest version)

- **Still plain `LinearRegression`.** Only the features/threshold changed.
- **The new features did NOT help.** On leak-free, multi-seed CV, the v1 lean 7-feature set
  ties or beats `+interactions` and `+cabin/ticket`. So we ship the lean set. This is a real
  result: `Family_Survival + Sex + Pclass` already capture what `Sex×Pclass` would add, and
  the extra columns mostly add noise. **More features ≠ better.**
- **The genuine win here is rigour, not score:**
  1. *Leak-free CV* — `Family_Survival` recomputed inside each fold ⇒ the ~0.84 CV is finally
     trustworthy (v1's 0.85 was inflated).
  2. *Stable threshold* — averaged over 5 seeds, the peak sits at ~0.54 (flat 0.53–0.58). A
     single split had suggested 0.60, which would have predicted too few survivors and likely
     scored *worse*. Robust tuning avoids that trap.
- **Honest expectation:** `submission1.csv` should score **about the same as v1 (~0.80)** —
  not a leap. We are at the real honest ceiling for this dataset. Beating it materially
  (0.82+) requires looking up the public answers, which the task explicitly tells us not to
  do. The value of v2 is *confidence that 0.80 is real and not luck.*
