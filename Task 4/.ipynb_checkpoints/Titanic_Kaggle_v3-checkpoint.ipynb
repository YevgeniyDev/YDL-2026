# Titanic — v3: regularized linear regression (Ridge / Lasso)

Goal: honestly try to pass **0.81339**. v1/v2 plateaued at 0.80143 with plain OLS on a
*lean* feature set — adding features overfit. **Ridge and Lasso are still linear regression**
(same `w0 + w1*x1 + ...` formula, same squared-error fit) but with a penalty that shrinks
coefficients:

```
OLS:    minimize  Σ(y - pred)^2
Ridge:  minimize  Σ(y - pred)^2 + alpha * Σ wi^2      (shrinks all weights)
Lasso:  minimize  Σ(y - pred)^2 + alpha * Σ |wi|      (shrinks + zeroes weak ones)
```

The penalty controls overfitting — which means we can now safely feed a **richer** feature
set (interactions, cabin, ticket-group, port, all titles) and let regularization decide what
matters. Everything is tuned on the same **leak-free, multi-seed CV** from v2, so the
comparison is honest. Output: **`submission2.csv`**.

> Note on the rule: Ridge/Lasso are *regularized linear regression*. Use this submission only
> if your interpretation of "only linear regression" allows regularization. The plain-OLS
> result (also computed below) remains available as the strict-rule fallback.

````python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score

train = pd.read_csv('train.csv'); test = pd.read_csv('test.csv')
n_train = len(train); y = train['Survived'].astype(float)
print('train:', train.shape, ' test:', test.shape)
````

## 1. Rich feature engineering (no labels used here)

Same base transforms as v2, plus we keep *every* feature this time — `Embarked` port, all
`Title` dummies, `HasCabin`, `TicketGroupSize`, and the `Sex x Pclass` / `Sex x AgeBin`
interactions. Under plain OLS this set overfits; under Ridge/Lasso it is fair game.

````python
combined = pd.concat([train, test], sort=False, ignore_index=True)

combined['RawTitle'] = combined['Name'].str.extract(r',\s*([^.]+)\.')[0].str.strip()
title_map = {'Mr':'Mr','Mrs':'Mrs','Miss':'Miss','Master':'Master','Ms':'Miss','Mlle':'Miss',
    'Mme':'Mrs','Lady':'Rare','Countess':'Rare','Dona':'Rare','Don':'Rare','Sir':'Rare',
    'Jonkheer':'Rare','Dr':'Rare','Rev':'Rare','Col':'Rare','Major':'Rare','Capt':'Rare'}
combined['Title'] = combined['RawTitle'].map(title_map).fillna('Rare')

combined['Sex_female'] = (combined['Sex'] == 'female').astype(int)
combined['FamilySize'] = combined['SibSp'] + combined['Parch'] + 1
combined['IsAlone']    = (combined['FamilySize'] == 1).astype(int)

combined['Age']  = combined['Age'].fillna(combined.groupby('Title')['Age'].transform('median'))
combined['Fare'] = combined['Fare'].fillna(combined.groupby('Pclass')['Fare'].transform('median'))
combined['Embarked'] = combined['Embarked'].fillna(combined['Embarked'].mode()[0])

combined['AgeBin']  = pd.cut(combined['Age'], bins=[0,14,25,35,45,60,200], labels=False)
combined['FareBin'] = pd.qcut(combined['Fare'], 5, labels=False)
combined['HasCabin']        = combined['Cabin'].notna().astype(int)
combined['TicketGroupSize'] = combined.groupby('Ticket')['Ticket'].transform('count')
combined['Sex_x_Pclass'] = combined['Sex_female'] * combined['Pclass']
combined['Sex_x_AgeBin'] = combined['Sex_female'] * combined['AgeBin']

combined = pd.get_dummies(combined, columns=['Title', 'Embarked'], dtype=int)
combined['LastName'] = combined['Name'].str.split(',').str[0]

rich_cols = [
    'Family_Survival','Sex_female','Pclass','AgeBin','FareBin','FamilySize','IsAlone',
    'HasCabin','TicketGroupSize','Sex_x_Pclass','Sex_x_AgeBin',
    'Title_Mr','Title_Mrs','Title_Miss','Title_Master','Title_Rare',
    'Embarked_C','Embarked_Q','Embarked_S',
]
print(len(rich_cols), 'features:', rich_cols)
````

## 2. Leak-safe `Family_Survival` + leak-free, scaled, multi-seed CV

`family_survival` is identical to v2 (reads only allowed labels, never a passenger's own).
The CV harness now also **fits a `StandardScaler` inside each fold** (needed so Ridge/Lasso
penalties treat every feature on the same scale) and accepts a model factory so we can score
OLS, Ridge(alpha), and Lasso(alpha) on exactly the same folds.

````python
def family_survival(df, known_mask):
    surv = df['Survived'].where(pd.Series(known_mask, index=df.index))
    fs = pd.Series(0.5, index=df.index)
    for _, grp in df.groupby(['LastName', 'Fare']):
        if len(grp) > 1:
            for idx in grp.index:
                o = surv.loc[grp.index.drop(idx)]
                if o.max() == 1.0:   fs.loc[idx] = 1.0
                elif o.min() == 0.0: fs.loc[idx] = 0.0
    for _, grp in df.groupby('Ticket'):
        if len(grp) > 1:
            for idx in grp.index:
                if fs.loc[idx] in (0.0, 0.5):
                    o = surv.loc[grp.index.drop(idx)]
                    if o.max() == 1.0:   fs.loc[idx] = 1.0
                    elif o.min() == 0.0: fs.loc[idx] = 0.0
    return fs

SEEDS = range(5)
seed_folds = {}
for seed in SEEDS:
    skf = StratifiedKFold(5, shuffle=True, random_state=seed)
    per_fold = []
    for tr_idx, va_idx in skf.split(np.arange(n_train), y):
        known = np.zeros(len(combined), bool); known[tr_idx] = True
        per_fold.append((tr_idx, va_idx, family_survival(combined, known)))
    seed_folds[seed] = per_fold

THRESHOLDS = np.arange(0.30, 0.71, 0.01)

def honest_oof(make_model, seed):
    oof = np.zeros(n_train)
    for tr_idx, va_idx, fs in seed_folds[seed]:
        combined['Family_Survival'] = fs.values
        X = combined[rich_cols].astype(float)
        sc = StandardScaler().fit(X.iloc[tr_idx])
        m = make_model().fit(sc.transform(X.iloc[tr_idx]), y.iloc[tr_idx])
        oof[va_idx] = m.predict(sc.transform(X.iloc[va_idx]))
    return oof

def mean_cv(make_model):
    acc = np.zeros(len(THRESHOLDS))
    for seed in SEEDS:
        oof = honest_oof(make_model, seed)
        acc += [accuracy_score(y, (oof >= t).astype(int)) for t in THRESHOLDS]
    acc /= len(SEEDS)
    i = int(acc.argmax())
    return THRESHOLDS[i], acc[i]
````

## 3. Compare OLS vs Ridge vs Lasso (with alpha search)

````python
candidates = [('OLS', lambda: LinearRegression())]
for a in [0.1, 0.3, 1.0, 3.0, 10.0, 30.0]:
    candidates.append((f'Ridge alpha={a}', (lambda a: lambda: Ridge(alpha=a))(a)))
for a in [0.001, 0.003, 0.01, 0.03, 0.1]:
    candidates.append((f'Lasso alpha={a}', (lambda a: lambda: Lasso(alpha=a, max_iter=10000))(a)))

scored = []
for name, mk in candidates:
    t, acc = mean_cv(mk)
    scored.append((name, mk, t, acc))
    print(f'{name:18s}  meanCV acc = {acc:.4f}  (threshold {t:.2f})')

best_name, best_mk, best_t, best_acc = max(scored, key=lambda r: r[3])
print(f'\nBest: {best_name}  ->  {best_acc:.4f} at threshold {best_t:.2f}')
````

## 4. Fit the winner on all data, write `submission2.csv`

````python
known_full = np.zeros(len(combined), bool); known_full[:n_train] = True
combined['Family_Survival'] = family_survival(combined, known_full).values

X      = combined.iloc[:n_train][rich_cols].astype(float)
X_test = combined.iloc[n_train:][rich_cols].astype(float).reset_index(drop=True)

scaler = StandardScaler().fit(X)
final  = best_mk().fit(scaler.transform(X), y)
pred   = (final.predict(scaler.transform(X_test)) >= best_t).astype(int)

pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': pred}).to_csv('submission2.csv', index=False)
print(f'Wrote submission2.csv using {best_name}')
print('Predicted survivors:', int(pred.sum()), 'of', len(pred), f'({pred.mean():.1%})')

print('\nCoefficients (scaled features):')
print(pd.Series(final.coef_, index=rich_cols).sort_values())
````

## 5. How different is this from v1's submission?

````python
import os
if os.path.exists('submission.csv'):
    v1 = pd.read_csv('submission.csv')['Survived'].values
    diff = int((v1 != pred).sum())
    print(f'submission2 differs from v1 on {diff} of {len(pred)} passengers')
    print('(~209 of these are on the public leaderboard; each flip is ~0.0048 score)')
````

## Summary

- **Method:** still linear regression, now in its **regularized** form (Ridge/Lasso). Same
  linear formula, just shrunk coefficients — chosen by the same leak-free multi-seed CV.
- **Why a richer feature set now works:** the penalty term controls overfitting, so the
  extra features (interactions, cabin, ticket-group, port, titles) can no longer blow up the
  variance the way they did under plain OLS in v2.
- **Honest expectation:** if Ridge/Lasso help, it is by a few passengers — exactly the
  ~0.012 that separates 0.80143 from 0.81339. That margin is mostly which side of the line a
  handful of borderline passengers fall on for this public split; it may not hold on the
  private leaderboard.
- **Rule reminder:** submit `submission2.csv` only if regularized linear regression counts as
  "linear regression" under your task's rules; otherwise keep v1's plain-OLS 0.80143.
