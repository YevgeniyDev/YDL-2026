# Можно ли верить вашей модели?
## Регрессия на реальных данных: от EDA до честного вердикта

**Сценарий.** Я junior-аналитик. Заказчик (ювелирный дом) приносит каталог из ~54 000
бриллиантов и просит: «Постройте модель, которая предсказывает **цену** камня по его
характеристикам, и скажите честно, насколько ей можно верить — можно ли по ней оценивать
камни для закупки».

**Что и для кого предсказываем (одна фраза).** Предсказываем `price` (цену бриллианта в \$)
по его весу и качеству (4C: carat, cut, color, clarity) и физическим размерам — чтобы
оценщик быстро прикидывал справедливую цену камня.

**Датасет:** Diamonds (53 940 бриллиантов, классический набор из `ggplot2`/`seaborn`).

> Линия дня: **не верь числу на слово — проверь.** Цель не красивый R², а доказать,
> где модель надёжна, а где врёт.

## 1. Загрузка данных и постановка задачи

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, learning_curve
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, SGDRegressor
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn import metrics

sns.set_theme(style="whitegrid")
RANDOM_STATE = 42

# --- Надёжная загрузка: локальный CSV, иначе через seaborn и кэшируем ---
import os
if os.path.exists("diamonds.csv"):
    df = pd.read_csv("diamonds.csv")
else:
    df = sns.load_dataset("diamonds")
    df.to_csv("diamonds.csv", index=False)  # кэш для офлайн-воспроизводимости

print("Размер:", df.shape)
df.head()
```

**Словарь признаков (ювелирные «4C» + геометрия):**

| признак | смысл | тип |
|---|---|---|
| `carat` | вес камня в каратах | непрерывный |
| `cut` | качество огранки | порядковый: Fair < Good < Very Good < Premium < Ideal |
| `color` | цвет (D — лучший, бесцветный … J — желтоватый) | порядковый: J … D |
| `clarity` | чистота | порядковый: I1 < SI2 < SI1 < VS2 < VS1 < VVS2 < VVS1 < IF |
| `depth`, `table` | пропорции огранки, % | непрерывные |
| `x`, `y`, `z` | размеры камня, мм | непрерывные |
| `price` | **ЦЕЛЬ:** цена, \$ | непрерывная |

Цель `price` непрерывная → это **регрессия**.

## 2. Разведочный анализ данных (EDA)

```python
df.info()
print("\nПропуски:", df.isnull().sum().sum())
```

```python
# Физически невозможные камни: нулевой ИЛИ абсурдно большой размер по x/y/z
print("Нулевые размеры:", {c: int((df[c] == 0).sum()) for c in ["x", "y", "z"]})
print("Размер > 20 мм (явные ошибки ввода):", {c: int((df[c] > 20).sum()) for c in ["x", "y", "z"]})
bad = (df[["x", "y", "z"]] == 0).any(axis=1) | (df[["x", "y", "z"]] > 20).any(axis=1)
print("Всего удаляем строк:", int(bad.sum()))
df = df[~bad].reset_index(drop=True)   # размер не может быть 0 или ~59 мм
print("После очистки:", df.shape)
df.describe()
```

**Ловушка №1 — брак в данных.** У части камней `x`/`y`/`z` равны 0 (камень не может иметь
нулевой размер), а у нескольких — абсурдные 30–59 мм (реальный бриллиант — единицы мм). Это
ошибки ввода. Их обязательно убрать: иначе один такой выброс через квадраты/кубы в полиноме
полностью дестабилизирует модель (мы это проверили — без очистки R² полинома «прыгает»).

### 2.1 Распределение цены

```python
fig, ax = plt.subplots(1, 2, figsize=(13, 4))
sns.histplot(df["price"], bins=60, kde=True, ax=ax[0])
ax[0].axvline(df["price"].mean(), color="red", ls="--", label=f"среднее={df['price'].mean():.0f}")
ax[0].axvline(df["price"].median(), color="green", ls="--", label=f"медиана={df['price'].median():.0f}")
ax[0].set_title("Цена price ($) — сильный правый перекос"); ax[0].legend()

sns.histplot(np.log(df["price"]), bins=60, kde=True, color="orange", ax=ax[1])
ax[1].set_title("log(price) — распределение почти симметрично")
plt.tight_layout(); plt.show()

print(f"Перекос (skew) цены: {df['price'].skew():.2f}  -> длинный правый хвост дорогих камней")
print(f"После log:           {np.log(df['price']).skew():.2f}")
```

**Ловушка №2 — скошенная цель.** Цена сильно скошена вправо (skew ≈ 1.6): среднее (≈3930\$)
почти вдвое больше медианы (≈2400\$), длинный хвост дорогих камней. RMSE будет «тянуться»
дорогими камнями; `log(price)` выпрямляет распределение — проверим это гипотезой позже.

### 2.2 Кодирование порядковых признаков и корреляции

```python
# cut/color/clarity имеют естественный порядок -> кодируем порядковыми числами
CUT = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
COLOR = ["J", "I", "H", "G", "F", "E", "D"]        # J (худший) ... D (лучший)
CLAR = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]

df["cut"] = df["cut"].map({k: i for i, k in enumerate(CUT)})
df["color"] = df["color"].map({k: i for i, k in enumerate(COLOR)})
df["clarity"] = df["clarity"].map({k: i for i, k in enumerate(CLAR)})

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Корреляционная матрица")
plt.tight_layout(); plt.show()

print("Корреляция признаков с ценой (по убыванию):")
print(df.corr(numeric_only=True)["price"].drop("price").sort_values(ascending=False).round(3))
```

**Находки по корреляциям.**
- Цену почти полностью определяет **размер**: `carat` (r ≈ 0.92) и габариты `x`, `y`, `z` (r ≈ 0.86–0.88).
- **Ловушка №3 — мультиколлинеарность.** `carat`, `x`, `y`, `z` измеряют по сути одно и то же
  (размер) и сильно скоррелированы между собой → коэффициенты линейной модели будут
  нестабильны, их нельзя интерпретировать «в лоб».
- Порядковые `cut`/`color`/`clarity` дают **слабую и даже отрицательную** линейную корреляцию с
  ценой — что выглядит парадоксально. Разберёмся ниже.

```python
fig, ax = plt.subplots(1, 2, figsize=(13, 4))
sample = df.sample(4000, random_state=RANDOM_STATE)
sns.scatterplot(data=sample, x="carat", y="price", hue="clarity", palette="viridis", alpha=0.5, ax=ax[0])
ax[0].set_title("carat vs price — связь НЕлинейная (цена растёт быстрее)")

# Парадокс чистоты: лучшая чистота -> ниже средняя цена (т.к. камни мельче)
g = df.groupby("clarity").agg(средняя_цена=("price", "mean"), средний_carat=("carat", "mean"))
g.index = CLAR
g["средняя_цена"].plot(kind="bar", ax=ax[1], color="steelblue")
ax[1].set_title("Средняя цена по чистоте (0=I1 хуже → 7=IF лучше)")
ax[1].set_ylabel("средняя цена, $")
plt.tight_layout(); plt.show()
print(g.round(2))
```

**Ловушка №4 — конфаундер (парадокс Симпсона).** Камни с *лучшей* чистотой (IF, VVS) в
среднем **дешевле**, чем с худшей (SI, I1)! Причина не в том, что чистота снижает цену, а в
том, что более чистые камни в выборке в среднем **мельче** (средний carat у IF ≈ 0.5 против
≈1.3 у I1), а вес важнее. Это классическая ловушка: при одинаковом весе лучшая чистота, конечно,
дороже — но «голая» корреляция это маскирует. Линейной модели нужно учесть это **совместно**.

### 2.3 Замеченные ловушки (итог EDA)

1. **Брак данных** — нулевые и абсурдно большие размеры x/y/z (убрали).
2. **Скошенная цель** — длинный хвост дорогих камней → кандидат на `log(price)`.
3. **Мультиколлинеарность** размерных признаков (carat, x, y, z) → коэффициенты неинтерпретируемы.
4. **Конфаундер качества** — чистота/цвет коррелируют с весом, поэтому «в одиночку» вводят в
   заблуждение.
5. **Нелинейность** связи carat→price (цена растёт быстрее веса) → полином должен помочь.

## 3. Базовая модель — честная оценка

Всё по-честному: разбиение train/test, масштабирование, `LinearRegression`,
**метрики только на тесте**.

```python
X = df.drop("price", axis=1)
y = df["price"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE)
print("train:", X_train.shape, " test:", X_test.shape)

baseline = Pipeline([
    ("scaler", StandardScaler()),
    ("lin_reg", LinearRegression()),
])
baseline.fit(X_train, y_train)
pred = baseline.predict(X_test)


def report(y_true, y_pred, name="модель"):
    mae = metrics.mean_absolute_error(y_true, y_pred)
    mse = metrics.mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = metrics.r2_score(y_true, y_pred)
    print(f"[{name}] МЕТРИКИ НА ТЕСТЕ:")
    print(f"  MAE  = {mae:.1f}")
    print(f"  MSE  = {mse:.1f}")
    print(f"  RMSE = {rmse:.1f}")
    print(f"  R2   = {r2:.3f}")
    return rmse, r2


report(y_test, pred, "Linear (baseline)");
```

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 4))
ax[0].scatter(y_test, pred, alpha=0.2)
lims = [y_test.min(), y_test.max()]
ax[0].plot(lims, lims, "r--")
ax[0].set_xlabel("Факт ($)"); ax[0].set_ylabel("Предсказание ($)")
ax[0].set_title("Предсказание vs факт")

residuals = y_test - pred
sns.histplot(residuals, bins=50, kde=True, ax=ax[1])
ax[1].axvline(0, color="red", ls="--")
ax[1].set_title("Распределение остатков")

ax[2].scatter(pred, residuals, alpha=0.2)
ax[2].axhline(0, color="red", ls="--")
ax[2].set_xlabel("Предсказание ($)"); ax[2].set_ylabel("Остаток ($)")
ax[2].set_title("Остатки vs предсказание (видна «воронка»)")
plt.tight_layout(); plt.show()
```

**Что говорят остатки базовой модели.** R² ≈ 0.91 — на первый взгляд отлично. Но честные
графики выдают проблемы:
- остатки vs предсказание имеют форму **«воронки»** (гетероскедастичность): чем дороже камень,
  тем сильнее разброс ошибки;
- модель даже **уходит в минус** для дешёвых камней (предсказывает отрицательную цену);
- остатки скошены — хвост дорогих камней недооценивается.

Это признаки того, что линейная форма + исходная (скошенная) шкала цены **недостаточны**. Будем
лечить нелинейность (полином) и перекос (`log`).

## 4. Улучшение и диагностика (методы главы 4)

На том же train/test split сравним линейную регрессию, полином и регуляризацию
(Ridge / Lasso / ElasticNet), проверим градиентный спуск и кривые обучения.

```python
models = {
    "Linear": Pipeline([
        ("scaler", StandardScaler()),
        ("reg", LinearRegression())]),
    "Poly(2)": Pipeline([
        ("poly", PolynomialFeatures(degree=2, include_bias=False)),
        ("scaler", StandardScaler()),
        ("reg", Ridge(alpha=1.0, random_state=RANDOM_STATE))]),
    "Poly(3)": Pipeline([
        ("poly", PolynomialFeatures(degree=3, include_bias=False)),
        ("scaler", StandardScaler()),
        ("reg", Ridge(alpha=1.0, random_state=RANDOM_STATE))]),
    "Ridge": Pipeline([
        ("scaler", StandardScaler()),
        ("reg", Ridge(alpha=1.0, random_state=RANDOM_STATE))]),
    "Lasso": Pipeline([
        ("scaler", StandardScaler()),
        ("reg", Lasso(alpha=1.0, random_state=RANDOM_STATE, max_iter=10000))]),
    "ElasticNet": Pipeline([
        ("scaler", StandardScaler()),
        ("reg", ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=RANDOM_STATE, max_iter=10000))]),
}

rows = []
for name, model in models.items():
    model.fit(X_train, y_train)
    p = model.predict(X_test)
    rows.append({
        "model": name,
        "test_RMSE": np.sqrt(metrics.mean_squared_error(y_test, p)),
        "test_R2": metrics.r2_score(y_test, p),
    })

results = pd.DataFrame(rows).set_index("model").sort_values("test_RMSE")
results.round(3)
```

**Чтение таблицы.**
- **Полином резко улучшает модель:** degree 2 поднимает R² 0.91 → ≈0.96, degree 3 → ≈0.98.
  Значит исходная проблема — именно **нелинейность** (цена растёт быстрее веса), и она лечится.
- **Ridge / Lasso** ≈ обычная линейная: на линейных признаках переобучения нет, регуляризовать
  почти нечего.
- **ElasticNet (alpha=1)** заметно **хуже** — это пример **избыточной регуляризации**: слишком
  сильный штраф «придавил» полезные коэффициенты и привёл к недообучению.

### 4.1 Перекрёстная проверка (k-fold CV)

Таблица выше построена на **одном** train/test split, а он может оказаться удачным или неудачным
(мы уже видели, как из-за выбросов R² полинома «прыгал» при смене разбиения). Поэтому подтверждаем
рейтинг моделей честной **5-блочной кросс-валидацией** на всех данных и смотрим не только среднее,
но и **разброс между фолдами**.

```python
from sklearn.model_selection import cross_validate, KFold

kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
cv_rows = []
for name, model in models.items():
    cv = cross_validate(model, X, y, cv=kf,
                        scoring=["neg_root_mean_squared_error", "r2"], n_jobs=-1)
    cv_rows.append({
        "model": name,
        "CV_RMSE": -cv["test_neg_root_mean_squared_error"].mean(),
        "RMSE_std": cv["test_neg_root_mean_squared_error"].std(),
        "CV_R2": cv["test_r2"].mean(),
        "R2_std": cv["test_r2"].std(),
    })
cv_results = pd.DataFrame(cv_rows).set_index("model").sort_values("CV_RMSE")
cv_results.round(3)
```

**Вывод по CV.** Рейтинг моделей тот же, что и на одиночном split, а разброс между фолдами
крошечный (станд. отклонение R² ≤ 0.004, RMSE ± ~20\$). Значит выбор в пользу полинома —
**устойчивый результат**, а не везение одного разбиения. Маленький std по фолдам ещё раз
подтверждает: переобучения нет, модель стабильна на разных подвыборках.

### 4.2 Проверка градиентного спуска (SGD) против аналитического решения

```python
sgd = Pipeline([
    ("scaler", StandardScaler()),
    ("reg", SGDRegressor(max_iter=2000, tol=1e-4, penalty=None,
                         eta0=0.01, random_state=RANDOM_STATE)),
])
sgd.fit(X_train, y_train)

print("Аналитическое решение (LinearRegression):")
report(y_test, baseline.predict(X_test), "Linear")
print("\nГрадиентный спуск (SGDRegressor):")
report(y_test, sgd.predict(X_test), "SGD")
print("\nВывод: SGD сходится примерно к тому же решению, что и нормальное уравнение.")
print("Масштабирование признаков критично: carat ~0.8, а x,y,z ~5-6 мм — без StandardScaler")
print("градиентный спуск сходился бы плохо.")
```

### 4.3 Честный эксперимент с log-таргетом

```python
# Лечим перекос: обучаем на log(price), ошибку считаем в исходных долларах (через exp)
best = Pipeline([
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("scaler", StandardScaler()),
    ("reg", Ridge(alpha=1.0, random_state=RANDOM_STATE)),
])
best.fit(X_train, np.log(y_train))
pred_log = np.exp(best.predict(X_test))   # обратно в доллары

rmse_log = np.sqrt(metrics.mean_squared_error(y_test, pred_log))
r2_log = metrics.r2_score(y_test, pred_log)
mae_log = metrics.mean_absolute_error(y_test, pred_log)
print(f"Poly(2) на log(price): RMSE={rmse_log:.1f}, R2={r2_log:.3f}, MAE={mae_log:.1f}")
print(f"Poly(2) без log:       RMSE={results.loc['Poly(2)','test_RMSE']:.1f}, "
      f"R2={results.loc['Poly(2)','test_R2']:.3f}")
print("\nLog-таргет особенно улучшает MAE и убирает отрицательные предсказания цены.")
```

### 4.4 Кривые обучения — диагноз

```python
# 54k строк: используем кросс-валидационный learning_curve из sklearn
# (поэлементный цикл из главы 4 был бы слишком медленным; смысл тот же).
def plot_lc(model, X, y, ax, title):
    sizes, train_sc, val_sc = learning_curve(
        model, X, y, cv=5, train_sizes=np.linspace(0.1, 1.0, 6),
        scoring="neg_root_mean_squared_error", n_jobs=-1)
    ax.plot(sizes, -train_sc.mean(axis=1), "r-+", lw=2, label="train RMSE")
    ax.plot(sizes, -val_sc.mean(axis=1), "b-", lw=2, label="validation RMSE")
    ax.set_xlabel("Размер обучающей выборки"); ax.set_ylabel("RMSE ($)")
    ax.set_title(title); ax.legend(); ax.set_ylim(0, 2000)


fig, ax = plt.subplots(1, 2, figsize=(14, 4))
plot_lc(models["Linear"], X, y, ax[0], "Linear — недообучение (high bias)")
plot_lc(models["Poly(2)"], X, y, ax[1], "Poly(2) — точнее, кривые сошлись")
plt.tight_layout(); plt.show()
```

**Диагноз по кривым обучения.**
- **Linear:** train- и validation-RMSE быстро сходятся к **высокому** уровню (~1200\$) и почти
  сливаются — классическое **недообучение (high bias)**. Добавлять данные бесполезно, нужна более
  гибкая модель (что и подтвердил полином).
- **Poly(2):** обе кривые сходятся к заметно **более низкому** RMSE при небольшом зазоре — модель
  поймала нелинейность и при этом не переобучается (на 54k строк данных хватает). Degree 3 даёт
  ещё ниже ошибку, но уже сложнее и ближе к переобучению — берём degree 2 как баланс.

### 4.5 Интерпретация коэффициентов (осторожно: мультиколлинеарность)

```python
lin = LinearRegression().fit(StandardScaler().fit_transform(X_train), y_train)
coef = (pd.DataFrame({"feature": X.columns, "coef": lin.coef_})
        .assign(abs_coef=lambda d: d["coef"].abs())
        .sort_values("abs_coef", ascending=False))
print("Коэффициенты на стандартизованных признаках:")
print(coef[["feature", "coef"]].to_string(index=False))
```

Из-за мультиколлинеарности размерных признаков коэффициенты «странные»: `carat` получает
огромный положительный вес, а `x`/`y`/`z` могут уходить в минус — модель перераспределяет один
и тот же «эффект размера» между скоррелированными признаками. **Вывод:** доверять знакам и
величинам отдельных коэффициентов здесь нельзя, хотя на качество предсказания это не влияет.
Зато видно, что `clarity` и `color` теперь (при учёте веса) дают положительный вклад — парадокс
из EDA снят.

## 5. Вердикт: можно ли верить модели?

```python
final_pred = pred_log          # лучшая модель: Poly(2) на log(price)
final_rmse = rmse_log

print("Выбранная модель: Полиномиальная регрессия degree=2 на log(price).")
report(y_test, final_pred, "Финальная: Poly(2)+log")
print()
print("=== RMSE В КОНТЕКСТЕ ЦЕЛИ ===")
print(f"Средняя цена       : {y.mean():.0f} $")
print(f"Ст. отклонение цены: {y.std():.0f} $")
print(f"RMSE модели        : {final_rmse:.0f} $")
print(f"RMSE / среднее      : {final_rmse / y.mean() * 100:.0f}%  (типичная ошибка ~ доля от средней цены)")
print(f"RMSE / std          : {final_rmse / y.std():.2f}  (<< 1 => модель сильно лучше, чем средняя цена)")
```

```python
# Где модель ошибается сильнее всего — 5 крупнейших остатков
err = pd.DataFrame({
    "факт_$": y_test,
    "предсказание_$": final_pred,
    "остаток_abs_$": (y_test - final_pred).abs(),
}).join(df[["carat", "cut", "color", "clarity"]])

print("ТОП-5 ошибок (где модель врёт сильнее всего):")
err.sort_values("остаток_abs_$", ascending=False).head(5).round(1)
```

### Честный вердикт

**Можно ли верить модели? — Да, для типичных камней; с осторожностью — для дорогих.**

- **Какую модель выбрали и почему.** Лучшая — **полиномиальная регрессия degree 2 на
  `log(price)`** (R² ≈ 0.97). Линейная модель недообучалась (R² ≈ 0.91, «воронка» остатков,
  кривые обучения упёрлись в высокий bias, отрицательные предсказания цены). Полином снял
  нелинейность связи carat→price, а `log` — перекос цены и заодно убрал отрицательные цены.
  Degree 3 чуть точнее по RMSE, но сложнее и ближе к переобучению — берём degree 2 как баланс
  точности и простоты. Регуляризация (Ridge/Lasso) выигрыша не дала (переобучения не было), а
  слишком сильный ElasticNet даже навредил.

- **RMSE в контексте.** RMSE ≈ 650\$ при средней цене ≈ 3930\$ (≈17%) и std ≈ 3990\$
  (RMSE/std ≈ 0.16). Модель в разы точнее, чем «угадывать среднюю цену», и пригодна для быстрой
  оценки камня.

- **Где модель надёжна / где врёт.** Надёжна на массовых камнях до ~5 000\$. Сильнее всего
  ошибается на **крупных и дорогих** камнях (хвост распределения): из-за гетероскедастичности
  абсолютная ошибка там в тысячах долларов — см. топ-5 остатков. Для таких камней модель — лишь
  ориентир, нужна ручная оценка эксперта.

- **Что улучшил бы.** Древесные модели (RandomForest / GradientBoosting) обычно дают по этому
  датасету R² ≈ 0.98 без ручного полинома и лучше ловят взаимодействия 4C; плюс стоит отдельно
  моделировать дорогой хвост или работать в `log`-шкале метрик.

> Итог по линии дня: высокий R² базовой модели (0.91) скрывал реальные проблемы —
> **график остатков и кривые обучения** показали недообучение и гетероскедастичность, а
> EDA вскрыла конфаундер чистоты. Честная проверка важнее красивого числа.

### Сколько камней попадает в «ненадёжную» зону (> \$5000)

Вердикт выше говорит, что модель надёжна примерно до \$5000. Посчитаем, какой доли камней это
касается и насколько хуже модель предсказывает дорогой сегмент.

```python
THRESH = 5000

# 1) Сколько дорогих камней во всём (очищенном) датасете
n_total = len(df)
n_exp = int((df["price"] > THRESH).sum())
print(f"Камней дороже ${THRESH}: {n_exp} из {n_total} "
      f"({n_exp / n_total * 100:.1f}% датасета) — это зона повышенного риска ошибки.")
print()

# 2) Точность ФИНАЛЬНОЙ модели отдельно по двум ценовым зонам (на тесте)
zone = pd.DataFrame({"факт": y_test.values, "pred": final_pred})
expensive = zone["факт"] > THRESH

rows = []
for label, mask in [("до $5000 (надёжно)", ~expensive), ("дороже $5000 (хуже)", expensive)]:
    g = zone[mask]
    rows.append({
        "зона": label,
        "камней_в_тесте": len(g),
        "доля_теста_%": round(len(g) / len(zone) * 100, 1),
        "MAE_$": round(metrics.mean_absolute_error(g["факт"], g["pred"]), 1),
        "RMSE_$": round(np.sqrt(metrics.mean_squared_error(g["факт"], g["pred"])), 1),
        "ср_ошибка_%_от_цены": round((g["факт"] - g["pred"]).abs().mean() / g["факт"].mean() * 100, 1),
    })
print(pd.DataFrame(rows).set_index("зона"))
```

**Вывод по сегментам.** Дорогих камней (> \$5000) — около **27% датасета**, то есть это не
редкий край, а заметная четверть рынка. На них **абсолютная** ошибка модели примерно в 4 раза
больше (RMSE ≈ \$1200 против ≈ \$280 на дешёвых). Но в **процентах от цены** ошибка в обеих зонах
близка (≈9%) — это заслуга `log`-таргета: модель ошибается пропорционально цене, а не на
фиксированную сумму. Практический смысл: для дорогих камней доверительный интервал в долларах
широкий (±тысячи \$), поэтому финальное слово — за экспертом-оценщиком, хотя относительная
точность остаётся приемлемой.
